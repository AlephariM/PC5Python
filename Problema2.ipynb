```python
# renombrado de columnas

# {'old_colname': 'new_colname'}
df_rename = df_winemag.rename(columns={'Unnamed: 0': 'index', 'country':'pais', 'points':'puntos', 'taster_twitter_handle':'taster_twitter', 'winery':'bodegas'})
df_rename.head(3)
```

```python
# Función para aplicar la transformación
def cambiar_US(valor):
    if valor == 'US':
        return 'United States of America'
    else:
        return valor

# Aplicar la función a la columna 'pais' y crear la nueva columna 'paisf'
df_rename['paisf'] = df_rename['pais'].apply(cambiar_US)

#PRIMER COLUMNA CREADA
df_rename.head(3)
```

```python
df_rename = df_rename.drop(['pais'], axis=1)
df_rename.head(3)
```

```python
url = 'https://gist.githubusercontent.com/kintero/7d1db891401f56256c79/raw/a61f6d0dda82c3f04d2e6e76c3870552ef6cf0c6/paises.csv'
df_pais_conti = pd.read_csv(url)
df_pais_conti.head()
```

```python
df_pais_conti.columns
```

```python
df_pais_conti = df_pais_conti.drop(['nombre'], axis=1)
df_pais_conti = df_pais_conti.drop([' nom', ' iso2', ' iso3', ' phone_code'], axis=1)
df_pais_conti=df_pais_conti.rename(columns={' name':'pais'})
#df_pais_conti.columns

df_pais_conti.head()
```

```python
df_combinado = pd.merge(df_rename, df_pais_conti, how="left", left_on="paisf", right_on="pais")
#SEGUNDA COLUMNA CREADA
df_combinado.head(3)
```

```python
def grado_precio(valor):
    if valor < 30:
        return 'barato'
    elif valor>= 30 and valor<60:
        return 'moderado'
    elif valor>=60:
        return 'caro'
    else:
        return 'no disponible'

df_combinado['rango_precio'] = df_combinado['price'].apply(grado_precio)
df_final = df_combinado.drop(['pais'], axis=1)

#TERCERA COLUMNA CREADA
df_final.head(3)
```

```python
# REPORTE POR AGRUPAMIENTO DE DATOS N°1

# Filtrar el DataFrame por continente
def mejores_vinos_por_continente(df, continente):
    condicion = df['continente'] == continente
    df_agrup1 = df[condicion]
    
    # Ordenar los vinos por puntuación en orden descendente
    df_agrup1 = df_agrup1.sort_values(by='puntos', ascending=False)
    
    return df_agrup1.head(3)



print("Mejores vinos en América:")
mejores_vinos_por_continente(df_final, 'América')
```

```python
# REPORTE AGRUPAMIENTO DE DATOS N°2

# Agrupar el DataFrame por país y calcular el promedio de precio y la suma de reviews
df_agrup2 = df_final.groupby(['paisf']).agg({'price': 'mean', 'puntos': 'sum'})

# Ordenar los resultados por el promedio de precio en orden descendente
df_agrup2 = df_agrup2.sort_values(by='puntos', ascending=False)

print("Promedio de precio de vino y cantidad de reviews por país (ordenado de mejor a peor):")
df_agrup2.head()
```

```python
# REPORTE AGRUPAMIENTO DE DATOS N°3

condicion = df_final['paisf'] == 'Peru'
df_agrup3 = df_final[condicion].sort_values('price', ascending=False)
df_agrup3.head(3)
```

```python
# REPORTE AGRUPAMIENTO DE DATOS N°4
condicion = df_final['puntos'] >= 70
df_agrup4 = df_final[condicion][ ['index', 'description', 'designation', 'puntos', 'price', 'paisf'] ] 
df_agrup4 = df_agrup4.sort_values('puntos', ascending=True)
df_agrup4
```

```python
# REPORTE AGRUPAMIENTO DE DATOS N°5

def seleccionar_mejor_vino(df_):
    # Ordenar el grupo por puntuación y luego por precio de manera descendente
    df_funcion = df_.sort_values(by=['puntos', 'price'], ascending=[False, False])
    # Seleccionar el primer vino (iloc para seleccionar la posicion 0 de esa lista)
    df_devolver = df_funcion.iloc[0]
    return df_devolver


df_agrup5 = df_final.sort_values(by='paisf')

df_agrup5 = df_agrup5.groupby('paisf').apply(seleccionar_mejor_vino)

df_agrup5 = df_agrup5[ ['description', 'puntos', 'price', 'variety', 'bodegas'] ]
df_agrup5.head()
```

```python
#Guardar en csv
df_agrup5.to_csv('data\mejores_vinos_por_pais.csv', sep= '|', encoding='utf-8', index=False)
#Guardar en excel
df_agrup5.to_excel('data\mejores_vinos_por_pais.xlsx', index='MejoresxPais')
```
